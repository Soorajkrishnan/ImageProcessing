# **Computer Vision using OpenCV**

```
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```
print(cv.__version__)
```

```
s=cv.imread('cat.jpg',1) #read_image to the program(image_location,0[greyscale]|1[RGB])
print(s)
```

```
cv.imwrite('cat1.png',s)#writing_image to a file:(filename,image_object)
```

```
cv2_imshow(s)#printing_image to the file
```

```
roi=s[40:150,40:220]#cropping image
cv2_imshow(roi)
```

```
grey=cv.cvtColor(s,cv.COLOR_BGR2GRAY)#color-converstion
cv2_imshow(grey)
```

```
np.size(s)
np.shape(s)
```

```
hist=cv.calcHist([s],[0],None,[256],[0,255])
plt.hist(s.ravel(),256,[0,256]);plt.show()
```
