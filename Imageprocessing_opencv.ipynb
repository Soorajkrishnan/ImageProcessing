```
canvas=np.zeros((300,300,3),dtype='uint8')#uint8(unsigned int 8bit): For creating a black image
green=(0,255,0)
white=(255,255,255)
red=(0,0,255)
blue=(255,0,0)
yellow=(0,255,255)
cv2.line(canvas,(0,0),(300,300),green,5)#used to draw a line on canvas: (file,(ycoordinate(row,col),xcoordinate(row,col)),color(BGR),thickness)
cv2.line(canvas,(0,150),(300,150),yellow,5)#used to draw a line on canvas
cv2.line(canvas,(150,300),(150,0),blue,5)#used to draw a line on canvas)

cv2.line(canvas,(300,0),(0,300),red,5)
cv2_imshow(canvas)
```

```
canvas2=np.zeros((300,300,3),dtype='uint8')
cv2.rectangle(canvas2,(100,200),(200,100),red,2)
cv2.rectangle(canvas2,(150,200),(200,150),white,2)
cv2.rectangle(canvas2,(300,150),(150,300),blue,2)
#cv2.rectangle(canvas2,(100,0),(0,100),red,2)
#cv2.rectangle(canvas2,(100,0),(0,100),red,2)
#cv2.rectangle(canvas2,(300,0),(200,100),white,2)
#cv2.rectangle(canvas2,(200,300),(100,0),blue,2)
cv2_imshow(canvas2)
```

```
canvas3=np.zeros((300,300,3),dtype='uint8')
grey=(220,220,220)
cv2.circle(canvas3,(150,150),50,grey,-1)
cv2_imshow(canvas3)
```

```
canvas4=np.zeros((300,300,3),dtype='uint8')
cv2.rectangle(canvas4,(25,25),(275,275),white,-1)
cv2_imshow(canvas4)
```

```
canvas5=np.zeros((300,300,3),dtype='uint8')
cv2.circle(canvas5,(150,150),150,white,-1)
cv2_imshow(canvas5)
```

```
bit_and=cv2.bitwise_and(canvas4,canvas5)
cv2_imshow(bit_and)
```

```
bit_or=cv2.bitwise_or(canvas4,canvas5)
cv2_imshow(bit_or)
```

```
bit_not1=cv2.bitwise_not(canvas5)
cv2_imshow(bit_not1)
```

```
bit_not2=cv2.bitwise_not(canvas4)
cv2_imshow(bit_not2)
```

```
bit_xor=cv2.bitwise_xor(canvas4,canvas5)
cv2_imshow(bit_xor)
```
